## Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

## Load Dataset

```python
path = "https://raw.githubusercontent.com/andynganle/Data-for-Assignment/master/nyc_subway.csv"
df = pd.read_csv(path)
```

From all the info above, we can guess that:
- **UNIT:** Id of the Subway
- **DATEn & TIMEn & datetime & hour:** Contain the information about the date and time of the Subway. The `DATEn & TIMEn` and `datetime & hour` actually contains the same information so we can choose to drop one pair of features.
- **ENTRIESn and EXITSn:** Number of Entries and Exits of the day
- **ENTRIESn_hourly and EXITSn_hourly:** Number of Entries and Exits of the hour
- **day_week:** Days of the week, value from 0 to 6
- **weekday:** The day is weekday or not? Value is 0 or 1
- **conds:** Weather Condition of that time of the day
- **fog:** Weather condition is fog or not, value is 0 or 1
- `precipi ~ weather_lon`: More info about weather.

## EDA

```python
df['ENTRIESn'].plot(kind='hist', bins=50, edgecolor='black')
```

```python
df['EXITSn'].plot(kind='hist', bins=50, edgecolor='black')
```

We can observe that the `ENTRIESn` and `EXITSn` have large skewness in the data

```python
# day_week vs ENTRIESn
plt.bar(df['day_week'], df['ENTRIESn'], width = 0.5)
plt.xlabel('day_week')
plt.ylabel('ENTRIESn')
```

```python
# day_week vs EXITSn
plt.bar(df['day_week'], df['EXITSn'], width = 0.5)
plt.xlabel('day_week')
plt.ylabel('EXITSn')
```

From the graph, we can observe that more people in NYC take the subway during weekdays than on the weekend.

```python
# plotting correlation heatmap

plt.figure(figsize = (20, 10))
dataplot = sns.heatmap(df.corr(), cmap="YlGnBu", annot=True)
# displaying heatmap
plt.show()
```

From the correlation map, we can see that the weather condition is not affect much people on taking the subway. Surprisingly, the `latitude` and `longtitude` has relative to `ENTRIESn` and `EXITSn`

```python
# conds vs ENTRIESn
plt.barh(df['conds'], df['ENTRIESn'], color='maroon')
```

```python
# conds vs EXITSn
plt.barh(df['conds'], df['EXITSn'], color='maroon')
```

More people take subway during Overcast and Clear than other weather condition.
